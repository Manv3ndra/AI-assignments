```python
class MyEightPuzzle:
    # constructor of the class
    def __init__(self, startState_102017093, goalState_102017093, startState2D_102017093, goalState2D_102017093):
        self.startState=startState_102017093
        self.goalState=goalState_102017093
        self.startState2D=startState2D_102017093
        self.goalState2D=goalState2D_102017093
    # calculates the euclidean distance
    def euclidean_distance_nirbhay(self):
        val=0
        for i in range(0, 3):
            for j in range(0, 3):
                goalNode_nirbhay=self.goalState2D[i][j]
                if goalNode_nirbhay==0:
                    continue
                goalIndexX=i
                goalIndexY=j
                for k in range(0, 3):
                    for l in range(0, 3):
                        startNode=self.startState2D[k][l]
                        if startNode==goalNode_nirbhay:
                            startIndexX=k
                            startIndexY=l
                            break
                x=abs(goalIndexX-startIndexX)
                y=abs(goalIndexY-startIndexY)
                distance=( pow(x, 2) + pow(y, 2))**0.5
                val+=distance
                print("Euclidean distance for " ,goalNode_nirbhay,"is" ,int(distance))
        print("Heuristic function value for Euclidean distance = ",val)
    # calculates the manhattan distance
    def manhattan_distance_nirbhay(self):
        val=0
        for i in range(0, 9):
            goalNode_102017093=self.goalState[i]
            if goalNode_102017093==0:
                continue
            goalIndex=i
            for j in range(0, 9):
                startNode=self.startState[j]
                if startNode==goalNode_102017093:
                    startIndex=j
                    break
            difference=abs(goalIndex-startIndex)
            if difference<3:
                moves=difference
            elif difference>=3 and difference<6:
                moves=difference%3 + 1
            elif difference>=6 and difference<8:
                moves=difference%3 + 2
            val+=moves
            print("Manhattan distance for",self.goalState[i],"is ",moves)
        print("Heuristic function value for Manhattan distance= ",val)
    # calculates minkowski distance
    def minkowski_distance_nirbhay(self):
        val=0
        p=float(input("Enter value of p: "))
        for i in range(0, 3):
            for j in range(0, 3):
                goalNode_nirbhay=self.goalState2D[i][j]
                if goalNode_nirbhay==0:
                    continue
                goalIndexX=i
                goalIndexY=j
                for k in range(0, 3):
                    for l in range(0, 3):
                        startNode=self.startState2D[k][l]
                        if startNode==goalNode_nirbhay:
                            startIndexX=k
                            startIndexY=l
                            break
                x=abs(goalIndexX-startIndexX)
                y=abs(goalIndexY-startIndexY)
                distance=pow( pow(x, p) + pow(y, p), 1/p )
                val+=distance
                print("Minkowski distance for" ,goalNode_nirbhay ,"is ",int(distance))
        print("Heuristic function value for Minkowski distance= ",val)
        
# input parameters
start=[2, 0, 3, 1, 8, 4, 7, 6, 5]
goal= [1, 2, 3, 8, 0, 4, 7, 6, 5]
start2D=[[2, 0, 3], [1, 8, 4], [7, 6, 5]]
goal2D= [[1, 2, 3], [8, 0, 4], [7, 6, 5]]
problem=MyEightPuzzle(start, goal, start2D, goal2D)
problem.euclidean_distance_nirbhay()
problem.manhattan_distance_nirbhay()
problem.minkowski_distance_nirbhay()
```

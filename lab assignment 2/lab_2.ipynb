```python
j1=int(input("Enter the capacity of jug 1: "))
j2=int(input("Enter the capacity of jug 2: "))
j3=int(input("Enter the capacity of jug 3: "))
aim=int(input("Enter the final amount: "))

def display():
  print("Select the rule:")
  print("Rule 1: Transfer whole content of jug 1 to jug 2")
  print("Rule 2: Transfer whole content of jug 1 to jug 3")
  print("Rule 3: Transfer whole content of jug 2 to jug 1")
  print("Rule 4: Transfer whole content of jug 2 to jug 3")
  print("Rule 5: Transfer whole content of jug 3 to jug 1")
  print("Rule 6: Transfer whole content of jug 3 to jug 2")
  print("Rule 7: Transfer water from jug 1 to jug 2 until jug 2 is full")
  print("Rule 8: Transfer water from jug 2 to jug 1 until jug 1 is full")
  print("Rule 9: Transfer water from jug 2 to jug 3 until jug 3 is full")
  print("Rule 10: Transfer water from jug 3 to jug 2 until jug 2 is full")
  print("Rule 11: Transfer water from jug 1 to jug 3 until jug 3 is full")
  print("Rule 12: Transfer water from jug 3 to jug 1 until jug 1 is full")
def rule(ch,x,y,z):

  if(ch==1):
    if(x>0 and (x+y)<=j2):
      return 0,x+y,z
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==2):
    if(x>0 and (x+z)<=j3):
      return 0,y,x+z
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==3):
    if(y>0 and (x+y)<=j1):
      return x+y,0,z
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==4):
    if(y>0 and (y+z)<=j3):
      return x,0,y+z
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==5):
    if(z>0 and (x+z)<=j1):
      return x+z,y,0
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==6):
    if(z>0 and (z+y)<=j2):
      return x,y+z,0
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==7):
    if(x>0 and (x+y)>=j2):
      return x-(j2-y),j2,z
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==8):
    if(y>0 and (x+y)>=j1):
      return j1,y-(j1-x),z
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==9):
    if(y>0 and (y+z)>=j3):
      return x,y-(j3-z),j3
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==10):
    if(z>0 and (z+y)>=j2):
      return x,j2,z-(j2-y)
    else:
      print("The rule could not be applied")
      return x,y,z
  elif(ch==11):
    if(x>0 and (x+z)>=j3):
      return x-(j3-z),y,j3
    else:
      print("The rule could not be applied")
      return x,y,z

  elif(ch==12):
    if(z>0 and (z+x)>=j1):
      return j1,z-(j1-x),z
    else:
      print("The rule could not be applied")
      return x,y,z

  else:
    print("The choice is invalid!!")

display()
x=12
y=z=0
while(True):
  if(x==aim):
    print("Goal state reached.")
    break
  else:
    ch=int(input("Enter the rule to be applied:"))
    x,y,z=rule(ch,x,y,z)
    print("Current status:",x,y,z)
```

#### Q3

```python
import copy
frontier = []
explored = []

def find_pos(current):
    for i in range(3):
        for j in range(3):
            if current[i][j] == 0:
                r,c = i, j
                return r,c

def up(current, r, c):
    if r != 0:
        child = copy.deepcopy(current)
        child[r][c],child[r - 1][c] = (child[r - 1][c],0)
        if child not in frontier and child not in explored:
            frontier.append(child)
            return child
    return None

def down(current, r, c):
    if r != 2:
        child = copy.deepcopy(current)
        child[r][c], child[r+1][c] = (child[r+1][c],0)
        if child not in frontier and child not in explored:
            frontier.append(child)
            return child
    return None


def left(current, r, c):
    if c != 0:
        child = copy.deepcopy(current)
        child[r][c], child[r][c-1] = (child[r][c-1],0)
        if child not in frontier and child not in explored:
            frontier.append(child)
            return child
    return None


def right(current, r, c):
    if c != 2:
        child = copy.deepcopy(current)
        child[r][c], child[r][c+1] = (child[r][c+1],0)
        if child not in frontier and child not in explored:
            frontier.append(child)
            return child
    return None


def is_goal(current, goal_state):
    if current == goal_state:
        return True
    else:
        return False

def search():
    initial_state = [[1, 2, 3], [8, 0, 4], [7, 6, 5]]
    goal_state = [[2, 8, 1], [0, 4, 3], [7, 6, 5]]
    frontier.append(initial_state)
    flag = 1
    while flag:
        current = frontier.pop(0)
        if is_goal(current,goal_state):
            flag = 0
            print(len(explored))
        else:
            r,c = find_pos(current)
            neighbor = [0] * 4
            neighbor[0] = up(current,r,c)
            neighbor[1] = down(current,r,c)
            neighbor[2] = left(current,r,c)
            neighbor[3] = right(current,r,c)
            explored.append(current)
            for i in neighbor:
                if i != None and i != 0:
                    if is_goal(i,goal_state):
                        flag = 0
                        print("Process:")
                        for state in explored:
                            print(state)
                        print("Total states reached:", len(explored))
                        print("Goal state reached.")

search()
```

```python
from sys import maxsize
from itertools import permutations
V = 4

def travellingSalesmanProblem(graph, s):
	vertex = []
	for i in range(V):
		if i != s:
			vertex.append(i)

	min_path = maxsize
	next_permutation=permutations(vertex)
	for i in next_permutation:

		current_pathweight = 0
		k = s
		for j in i:
			current_pathweight += graph[k][j]
			k = j
		current_pathweight += graph[k][s]
		min_path = min(min_path, current_pathweight)
		
	return min_path

if __name__ == "__main__":
	graph = [[0, 10, 15, 20], [10, 0, 35, 25],
			[15, 35, 0, 30], [20, 25, 30, 0]]
	s = 0
	print(travellingSalesmanProblem(graph, s))
```

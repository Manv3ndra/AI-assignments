#### Q1. Given two jugs- a 4 litre and 3 litre capacity. Neither has any measurable markers on it. There is a pump which can be used to fill the jugs with water. Simulate the procedure in Python to get exactly 2 litre of water into 4-litre jug

#### Q3. Write a code in python for the 8 puzzle problem by taking the following initial and final states


```python
import copy
latestState = []
exploredState = []

def findBlankTile(currentState):
    for i in range(3):
        for j in range(3):
            if currentState[i][j] == 0:
                r,c = i, j
                return r,c

def moveUp(currentState, r, c):
    if r != 0:
        child = copy.deepcopy(currentState)
        child[r][c],child[r - 1][c] = (child[r - 1][c],0)
        if child not in latestState and child not in exploredState:
            latestState.append(child)
            return child
    return None

def moveDown(currentState, r, c):
    if r != 2:
        child = copy.deepcopy(currentState)
        child[r][c], child[r+1][c] = (child[r+1][c],0)
        if child not in latestState and child not in exploredState:
            latestState.append(child)
            return child
    return None


def moveLeft(currentState, r, c):
    if c != 0:
        child = copy.deepcopy(currentState)
        child[r][c], child[r][c-1] = (child[r][c-1],0)
        if child not in latestState and child not in exploredState:
            latestState.append(child)
            return child
    return None


def moveRight(currentState, r, c):
    if c != 2:
        child = copy.deepcopy(currentState)
        child[r][c], child[r][c+1] = (child[r][c+1],0)
        if child not in latestState and child not in exploredState:
            latestState.append(child)
            return child
    return None


def is_goal(currentState, goal_state):
    if currentState == goal_state:
        return True
    else:
        return False

def search():
    initial_state = [[1, 2, 3], [8, 0, 4], [7, 6, 5]]
    goal_state = [[2, 8, 1], [0, 4, 3], [7, 6, 5]]
    latestState.append(initial_state)
    flag = 1
    while flag:
        currentState = latestState.pop(0)
        if is_goal(currentState,goal_state):
            flag = 0
            print(len(exploredState))
        else:
            r,c = findBlankTile(currentState)
            neighbor = [0] * 4
            neighbor[0] = moveUp(currentState,r,c)
            neighbor[1] = moveDown(currentState,r,c)
            neighbor[2] = moveLeft(currentState,r,c)
            neighbor[3] = moveRight(currentState,r,c)
            exploredState.append(currentState)
            for i in neighbor:
                if i != None and i != 0:
                    if is_goal(i,goal_state):
                        flag = 0
                        print("Process:")
                        for state in exploredState:
                            print(state)
                        print("Total states reached:", len(exploredState))
                        print("Goal state reached.")

search()
```

#### Q4. Travelling Salesman Problem

```python
from sys import maxsize
from itertools import permutations
V = 4

def travellingSalesmanProblem(graph, s):
    vertex = []
    for i in range(V):
        vertex.append(i)

    min_path = maxsize
#     print(min_path)
#     print(vertex)
    next_permutation=permutations(vertex)
    for i in next_permutation:
        print(i)
        current_pathweight = 0
        k = s
        for j in i:
            current_pathweight += graph[k][j]
            k = j
        
#         print('k={}, s={}, graph[k][s]={}'.format(k,s,graph[k][s]))
        current_pathweight += graph[k][s]
        print(current_pathweight)
        min_path = min(min_path, current_pathweight)

    return min_path


graph = [[0, 10, 15, 20], [10, 0, 35, 25],
        [15, 35, 0, 30], [20, 25, 30, 0]]
s = 0
print(travellingSalesmanProblem(graph, s))
```
